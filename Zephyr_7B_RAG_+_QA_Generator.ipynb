```python

import torch
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

model_name = "HuggingFaceH4/zephyr-7b-alpha"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto"
)

qa_model = pipeline("text-generation", model=model, tokenizer=tokenizer,
                    max_new_tokens=256, do_sample=False, temperature=0.2, top_p=0.95,
                    return_full_text=False)
```

```python
def generate_prompt(text, num_qs, difficulty=3, domain='general', topic=None, output_format="QA", mode="basic"):
    topic_info = f" on the topic of {topic}" if topic else ""
    domain_context = f"You are an expert in {domain}{topic_info}."
    difficulty_info = f"Target difficulty level: {difficulty} on a 1–5 scale."

    formats = {
        "QA": "Generate {num_qs} Question-Answer pairs. Format: Q1: ..., A1: ...",
        "MCQ": "Generate {num_qs} Multiple-Choice Questions with 4 options (A, B, C, D). Mark the correct answer clearly.",
        "TF": "Generate {num_qs} True/False Questions. Format: Q1: ..., Answer: True/False.",
        "Short": "Generate {num_qs} Short-Answer Questions. Format: Q1: ..., Answer: ..."
    }

    instruction = formats.get(output_format, formats["QA"]).format(num_qs=num_qs)

    if mode == "basic":
        prompt = f"""{domain_context}

Please create high-quality questions based on the text provided.

{text}

{instruction}

{difficulty_info}
"""
    elif mode == "blooms":
        prompt = f"""{domain_context}

Based on Bloom's Taxonomy, generate {num_qs} questions targeting higher-order thinking (Application, Analysis, Evaluation).

{text}

{instruction}

{difficulty_info}
"""
    elif mode == "rare":
        prompt = f"""{domain_context}

Apply RARE (Retrieve-Ask-Read-Explain) prompting strategy to first retrieve key facts, then ask a focused question, read carefully, and explain clearly.

Generate {num_qs} questions accordingly.

{text}

{instruction}

{difficulty_info}
"""
    else:
        raise ValueError("Invalid mode selected. Choose 'basic', 'blooms', or 'rare'.")

    return prompt
```

```python
# Install if needed
!pip install pdfplumber

import pdfplumber
import fitz
from langchain.text_splitter import RecursiveCharacterTextSplitter

def extract_text_from_pdf(path):
    doc = fitz.open(path)
    all_text = ""
    for page in doc:
        text = page.get_text()
        all_text += text + "\n"
    return all_text

def extract_text_from_txt(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read().strip()

def chunk_text(text, chunk_size=800, overlap=100):
    splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=overlap)
    return splitter.split_text(text)
```

```python

import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

embed_model = SentenceTransformer("all-MiniLM-L6-v2")

def embed_chunks(chunks):
    embeddings = embed_model.encode(chunks)
    dim = embeddings.shape[1]
    index = faiss.IndexFlatL2(dim)
    index.add(np.array(embeddings))
    return index, embeddings, chunks

def retrieve_chunks(query, index, chunks, k=3):
    q_vec = embed_model.encode([query])
    _, I = index.search(q_vec, k)
    return [chunks[i] for i in I[0]]
```

```python

from google.colab import files
uploaded = files.upload()

filename = list(uploaded.keys())[0]
ext = filename.split('.')[-1]

if ext == "pdf":
    raw_text = extract_text_from_pdf(filename)
elif ext == "txt":
    raw_text = extract_text_from_txt(filename)

else:
    raise ValueError("Unsupported file type. Use .pdf or .txt")

chunks = chunk_text(raw_text)
index, embeddings, source_chunks = embed_chunks(chunks)
print(f"✅ Uploaded '{filename}' with {len(source_chunks)} chunks.")
```

```python

# User query and parameters
user_query = "Generate"
num_qs = int(input("\nHow many questions do you want to generate? (e.g., 5, 10): "))
difficulty = int(input("\nChoose difficulty (1=Easy, 5=Hard): "))
output_format = input("\nChoose output format (QA / MCQ / TF / Short): ").strip().upper()
mode = input("\nChoose mode (basic / blooms / rare): ").strip().lower()
domain = input("\nEnter domain (e.g., healthcare, technology, education): ") or 'general'
topic = input("\nOptional: Enter specific topic if any (or leave blank): ") or None

print(user_query,"\n", index, "\n",source_chunks)

retrieved = retrieve_chunks(user_query, index, source_chunks, k=3)
context = "\n".join(retrieved)
prompt = generate_prompt(context, num_qs, difficulty, domain, topic, output_format,mode)
print(prompt)
response = qa_model(prompt)[0]['generated_text']
print("\n=== Generated Questions & Answers ===\n")
print(response)
```
